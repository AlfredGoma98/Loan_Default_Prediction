### The dataset has:

- Categorical columns stored as str
- Binary Yes/No columns stored as str
- Numeric columns already correct  
We convert:
- Education → category
- EmploymentType → category
- MaritalStatus → category
- LoanPurpose → category
- HasMortgage, HasDependents, HasCoSigner → binary 0/1

```python
# Converting categorical columns to category data type
categorical_cols = [
    'Education','EmploymentType','MaritalStatus','LoanPurpose'
]

for col in categorical_cols:
    df[col] = df[col].astype('category')
```

```python
# Converting Yes/No columns to binary
binary_cols = ['HasMortgage','HasDependents','HasCoSigner']

for col in binary_cols:
    df[col] = df[col].map({'Yes':1, 'No':0})
```

```python
# Dropping duplicates
df = df.drop_duplicates()
```

```python
# saving the cleaned dataset
df.to_csv("C:/Users/user/Loan_default_predictor/data/clean/Loan_default_clean.csv", index=False)
```
